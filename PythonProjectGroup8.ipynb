```python
def time_of_day(t):
    if t in range(6,12):
        return 'Morning Hours'
    elif t in range(12,16):
        return 'Afternoon Hours'
    elif t in range(16,22):
        return 'Evening Hours'
    else:
        return 'Late Night Hours'
```

```python
yellow_data['pickup_hourofday']=yellow_data['tpep_pickup_datetime'].apply(time_of_day)
yellow_data['dropoff_hourofday']=yellow_data['tpep_dropoff_datetime'].apply(time_of_day)
```

```python
def payment_type(p):
    if p == 1:
        return 'Credit Card'
    elif p == 2:
        return 'Cash'
    elif p == 3:
        return 'No Charge'
    elif p == 4:
      return 'Dispute'
    else:
        return 'Voided Trip'
```

```python
def rate_code(p):
    if p == 1:
        return 'Standard Rate'
    elif p == 2:
        return 'JFK'
    elif p == 3:
        return 'Newark'
    elif p == 4:
      return 'Nassau or Westchester'
    elif p == 5:
      return 'Negotiated Fare'
    elif p == 6:
      return 'Group Ride' 
    else:
        return 'Not Defined'
```

```python
green_data['payment']=green_data['payment_type'].apply(payment_type)
green_data['payment']=green_data['payment_type'].apply(payment_type)

yellow_data['payment']=yellow_data['payment_type'].apply(payment_type)
yellow_data['payment']=yellow_data['payment_type'].apply(payment_type)
```

```python
class yellow_taxi():
  def __init__(self):
    self.tip_amount = 0
  def average_tipsy():
    total_tipsy = yellow_data['tip_amount'].sum()
    total_tripsy = len(yellow_data.index) - 1
    avg_tips = total_tipsy / total_tripsy
    print('Average tip amount for yellow taxis is: {}'.format(avg_tips))
  def histo():
    fig = px.histogram(yellow_data, x='payment')
    fig.show()

  def box():
    fig = px.box(yellow_data, y='pickup_hour')
    fig.show()

  def hist():
    fig = px.histogram(yellow_data, x='fare_amount')
    fig.show()

  def countplotven():
    sns.countplot(x ='VendorID', data = yellow_data)

  def countpass():
    sns.countplot(x ='passenger_count', data = yellow_data)

  def barplotfare():
    sns.barplot(y ='fare_amount', x ='pickup_hour' , data = yellow_data)

  def relplot():
    sns.relplot(x ="fare_amount", y ="tip_amount",
    data = yellow_data);

  def displot():
    sns.displot(yellow_data['pickup_hour'], kde =True, bins = 25)

  def catplot():
    sns.catplot(y='fare_amount',x='passenger_count',data=yellow_data,kind='strip')

  def lineplot():
    sns.lineplot(y='fare_amount',x='pickup_hour',data=yellow_data)

  def pairplot():
    sns.pairplot(yellow_data)

  def barplotcon():
    sns.barplot(y ='congestion_surcharge', x ='pickup_hour' , data = yellow_data)

  def barplottip():
    sns.barplot(y ='tip_amount', x ='pickup_hour' , data = yellow_data)

  def displot():
    sns.displot(yellow_data['tip_amount'], kde =True, bins = 25)
    
  def heatmap():
    yellow_pt = yellow_data.pivot_table(index='pickup_hour', columns='trip_distance',values='fare_amount')
    sns.heatmap(yellow_pt)

  def heatmapco():
    yellow = yellow_data.corr()
    sns.heatmap(yellow, annot=True, cmap='coolwarm', linecolor = 'black', linewidths=1 )
  
  
  
  

  

  

  
```
